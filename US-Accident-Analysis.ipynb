```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format="%Y-%m-%d %H:%M:%S", errors='coerce')
```

```python
df.Start_Time[0]
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde = False, norm_hist=True)
```

From the above graph we can see that a higher percentage of accident occurs between 6am to 10pm and 3pm to 6pm.

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde = False, norm_hist=True)
```

From the above graph we can see that most accidents occurs on weekdays and less on weekends. 

```python
sunday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sunday_start_time.dt.hour, bins=24, kde = False, norm_hist=True)
```

```python
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday_start_time.dt.hour, bins=24, kde = False, norm_hist=True)
```

From the above graphs we can see that on sundays the most accident occurs between 11am to 4pm, unlike weekdays.

```python
sns.distplot(df.Start_Time.dt.month, bins=12, kde = False, norm_hist=True)
```

From the above graph we can see that most of the accidents occurs in December.

```python
df_month_check = df[df.Start_Time.dt.year == 2023]
sns.distplot(df_month_check.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

```python
threshold = 86
df['Temp_category'] = df['Temperature(F)'].apply(lambda x: 'Warmer' if x >= threshold else 'Cooler')
```

```python
accident_temp = df['Temp_category'].value_counts()
```

```python
accident_temp
```

```python
accident_temp.plot(kind='bar', color=['blue','orange'])
plt.xticks(rotation=0)
```

From the above graph we can see that most of the accidents occur in Cooler Areas or in Winter Season.
